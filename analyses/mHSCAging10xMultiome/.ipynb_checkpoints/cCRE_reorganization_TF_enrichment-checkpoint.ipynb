### Load input data

```r
# Load differential nucleosome results
diffNuc <- read.table(
  "/home/yanhu/data/PRINT/multiScaleFootprinting/data/mHSCAging10xV3/diff_nucleosomes.tsv",
  sep = "\t", header = T, row.names = 1
)

# Calculate FDRs
diffNuc$score_fdr <- p.adjust(diffNuc$score_pval, method = "fdr")
diffNuc$pos_fdr <- p.adjust(diffNuc$pos_pval, method = "fdr")

# Load cCRE ranges
regions <- readRDS("/home/yanhu/data/PRINT/multiScaleFootprinting/data/mHSCAging10xV3/regionRanges.rds")

# Load TF motif matches
TFMotifRanges <- readRDS("/home/yanhu/data/PRINT/multiScaleFootprinting/data/mHSCAging10xV3/TFMotifRanges.rds")
```

```r
# Convert to a GRanges object
diffNucRanges <- stringr::str_split_fixed(rownames(diffNuc), "[:-]", 4)
diffNucRanges <- GRanges(
    seqnames = diffNucRanges[, 1], 
    ranges = IRanges(
        start = as.integer(diffNucRanges[, 2]) + as.integer(diffNucRanges[, 4]),
        end = as.integer(diffNucRanges[, 2]) + as.integer(diffNucRanges[, 4])
    )
)
diffNucRanges <- resize(diffNucRanges, 1000, fix = "center")
```

```r
# Whether to only focus on regions that are not differentially accessible during aging
nonDiffATAC <- F
if(nonDiffATAC){
  keptNuc <- 1:length(diffNucRanges)
  for(subpop in c("Mk-biased", "multi-lineage")){
    diffATAC <- read.table(paste0("../../data/mHSCAging10xV3/diffATAC_", subpop, ".tsv"))
    diffATAC <- diffATAC[(diffATAC$padj > 0.1),]
    keptNuc <- intersect(keptNuc, findOverlaps(
      GRanges(rownames(diffATAC)), 
      resize(diffNucRanges, 1, fix = "center")
    )@to)
  }
  diffNuc <- diffNuc[keptNuc, ]
  diffNucRanges <- diffNucRanges[keptNuc, ]
}
```

### Comput motif enrichment in regions with differential nucleosomes

```r
# Create RangedSummarizedExperiment object for footprint data
nucSE <- SummarizedExperiment(
    list(counts = as.matrix(cbind(diffNuc$mean_score, diffNuc$mean_score))), # ChromVAR does not accept a single column so we make a fake second column
    rowRanges = diffNucRanges
)
rownames(nucSE) <- as.character(rowRanges(nucSE))

# Add GC bias
nucSE <- chromVAR::addGCBias(nucSE, genome=BSgenome.Mmusculus.UCSC.mm10::BSgenome.Mmusculus.UCSC.mm10)

# Impute NA GC bias values with the mean of the rest of the values
bias <- nucSE@rowRanges$bias
nucSE@rowRanges$bias[is.na(bias)] <- mean(bias[!is.na(bias)])

# Find background peaks with matched GC content and accessibility
bgPeaks <- chromVAR::getBackgroundPeaks(nucSE,niterations=100) 
```

```r
# Find motif matches
motifMatches <- pbmcapply::pbmcmapply(
  function(TF){
    matchedInds <- unique(findOverlaps(rowRanges(nucSE), TFMotifRanges[[TF]])@from)
    matchVec <- rep(0, dim(nucSE)[1])
    matchVec[matchedInds] <- 1
    matchVec
  },
  names(TFMotifRanges),
  mc.cores = 8
)
colnames(motifMatches) <- names(TFMotifRanges)
```

```r
# Find regions contaning differential nucleosomes as foreground
filter <- (diffNuc$mean_diff_score < -1) & (diffNuc$score_fdr < 0.01)
fgRegions <- unique(findOverlaps(rowRanges(nucSE), diffNucRanges[filter])@from)
```

```r
enrichment <- BuenRTools::motifPeakZtest(
  peakSet = fgRegions,
  bgPeaks = bgPeaks,
  tfMat = motifMatches 
)
```

```r
paste(rownames(enrichment[order(-enrichment$z_test), ])[1:50], collapse = " ")
```

### Visualize example regions

```r
old_TFBS_bw <- rtracklayer::import.bw("/home/yanhu/data/PRINT_rev/mouse_HSC/Old_Mk_TFBS.bigwig")
young_TFBS_bw <- rtracklayer::import.bw("/home/yanhu/data/PRINT_rev/mouse_HSC/Young_Mk_TFBS.bigwig")
```

```r
TF <- "Yy1"

# Find regions contaning differential nucleosomes as foreground
filter <- (diffNuc$mean_diff_score < -1) & (diffNuc$score_fdr < 0.01)

# Keep motif sites around such nucleosomes
motifSites <- subsetByOverlaps(TFMotifRanges[[TF]], diffNucRanges[filter])

# For each age group, calculate the TF binding score for each motif site
TFBSScores <- list()
for(age in c("Old", "Young")){
    if(age == "Old"){
        bigwig <- old_TFBS_bw
    }else{
        bigwig <- young_TFBS_bw
    }
    ov <- findOverlaps(motifSites, bigwig)
    ovMat <- Matrix::sparseMatrix(
        i = ov@from,
        j = ov@to,
        x = 1, dims = c(length(motifSites), length(bigwig))
    )
    scores <- (ovMat %*% bigwig$score) 
    TFBSScores[[age]] <- scores[, 1] / width(motifSites) # Convert a n-by-1 matrix to vector
}

# Get the difference across aging
diffScores <- TFBSScores[["Old"]] - TFBSScores[["Young"]]
```

```r
selectedSites <- motifSites[(diffScores < -0.075) & (TFBSScores[["Young"]] < 0.4)]
findOverlaps(selectedSites, regions)@to
```

```r
options(repr.plot.width = 10, repr.plot.height = 8)
qplot(TFBSScores$Young, diffScores)
```

```r
region <- regions[regionInd]
bins <- tile(region, width = 1)[[1]]

# Retrieve TFBS scores in young HSCs in the current region
young_region_TFBS <- rep(0, length(bins))
ov <- findOverlaps(bins, young_TFBS_bw)
young_region_TFBS[ov@from] <- young_TFBS_bw[ov@to]$score

# Retrieve TFBS scores in old HSCs in the current region
old_region_TFBS <- rep(0, length(bins))
ov <- findOverlaps(bins, old_TFBS_bw)
old_region_TFBS[ov@from] <- old_TFBS_bw[ov@to]$score
```

```r
library(patchwork)
options(repr.plot.width = 12, repr.plot.height = 6)
trackList <- list(
    "Young" = young_region_TFBS,
    "Old" = old_region_TFBS,
    "Diff" = old_region_TFBS - young_region_TFBS
)
plotList <- list()
for(mode in names(trackList)){
    track <- trackList[[mode]]
    plotData <- data.frame(
        position = 1:length(track),
        score = track
    )
    plotList[[mode]] <- ggplot(plotData) +
        geom_line(aes(x = position, y = score)) +
        ylab(paste(mode, "\n TFBS")) + xlab("Position")
}
Reduce("/", plotList)
```
