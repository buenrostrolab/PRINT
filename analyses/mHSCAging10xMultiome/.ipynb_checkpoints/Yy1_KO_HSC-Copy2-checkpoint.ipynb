```r
library(tximport)
library(DESeq2)
library(ggplot2)
```

```r
# List the quantification files from Salmon: one quant.sf file per sample
aligned_dir <- "/home/yanhu/data/PRINT/multiScaleFootprinting/data/mHSCAging10xV3/Yy1_KO_Bcell/aligned"
salmon_files <- dir(aligned_dir, recursive=TRUE, pattern="quant.sf", full.names=TRUE)
names(salmon_files) <- gsub("_quant.sf", "", dir(aligned_dir))

# Read in the two-column data.frame linking transcript id (column 1) to gene id (column 2)
tx2genes <- read.table("/home/yanhu/utils/genomes/tx2gene.gencode.vM25.tsv", sep="\t", header=F)
colnames(tx2genes) <- c("transcript_id", "gene_id", "gene_name")

# tximport can import data from Salmon, Kallisto, Sailfish, RSEM, Stringtie
# here we summarize the transcript-level counts to gene-level counts
txi <- tximport(salmon_files, type = "salmon", tx2gene = tx2genes)
```

```r
# Prepare sample metadata table
sample_table <- data.frame(
    ID = c("SRR2532636", "SRR2532637", "SRR2532638", 
           "SRR2532639", "SRR2532640", "SRR2532641"),
    sample = c("Yy1_WT_rep_1", "Yy1_WT_rep_2", "Yy1_WT_rep_3",
              "Yy1_KO_rep_1", "Yy1_KO_rep_2", "Yy1_KO_rep_3"),
    condition = c(rep("WT", 3), rep("KO", 3))
)
rownames(sample_table) <- sample_table$ID
```

```r
# Create a DESeq2 object based on Salmon per-gene counts
dds <- DESeqDataSetFromTximport(txi, colData = sample_table, design = ~ condition)

# Model fitting
dds <- DESeq2::DESeq(dds)

# Retrieve differential analysis results
yy1_diffRNA <- DESeq2::results(dds, contrast = c("condition","KO","WT"))
yy1_diffRNA <- yy1_diffRNA[!is.na(yy1_diffRNA$padj),]
yy1_diffRNA <- yy1_diffRNA[order(sign(yy1_diffRNA$log2FoldChange) * -log10(yy1_diffRNA$padj),
                         decreasing = T),]

# Replace gene ids with gene names
rownames(yy1_diffRNA) <- tx2genes$gene_name[match(rownames(yy1_diffRNA), tx2genes$gene_id)]
```

```r
# Load motif match locations of de-novo motifs
TFMotifSites <- data.table::fread(
    "/data/yanhu/PRINT_rev/mouse_HSC/denovo_hits.bed",
    showProgress = T, nThread = 8
)

# Find the unique set of de-novo motifs
uniqueTFs <- unique(TFMotifSites$tf)

# Convert to a GRanges object
TFMotifRanges <- pbmcapply::pbmclapply(
    uniqueTFs,
    function(TF){
        sites <- TFMotifSites[TFMotifSites$tf == TF, ]
        GRanges(seqnames = sites$chr, ranges = IRanges(start = sites$start, end = sites$end))
    },
    mc.cores = 8
)
names(TFMotifRanges) <- uniqueTFs

TF_sites <- TFMotifRanges[["Thap1,Mbtps2,Yy1,Yy2,Taf1,Zfp42"]]
```

```r
old_TFBS_bw <- rtracklayer::import.bw("/home/yanhu/data/PRINT_rev/mouse_HSC/Old_Mk_TFBS.bigwig")
young_TFBS_bw <- rtracklayer::import.bw("/home/yanhu/data/PRINT_rev/mouse_HSC/Young_Mk_TFBS.bigwig")
```

```r
# Load differential nucleosome results
diff_nuc <- read.table(
  "/home/yanhu/data/PRINT/multiScaleFootprinting/data/mHSCAging10xV3/diff_nucleosomes.tsv",
  sep = "\t", header = T, row.names = 1
)

# Calculate FDRs
diff_nuc$score_fdr <- p.adjust(diff_nuc$score_pval, method = "fdr")
diff_nuc$pos_fdr <- p.adjust(diff_nuc$pos_pval, method = "fdr")

# Convert to a GRanges object
diff_nuc_ranges <- stringr::str_split_fixed(rownames(diff_nuc), "[:-]", 4)
diff_nuc_ranges <- GRanges(
    seqnames = diff_nuc_ranges[, 1], 
    ranges = IRanges(
        start = as.integer(diff_nuc_ranges[, 2]) + as.integer(diff_nuc_ranges[, 4]),
        end = as.integer(diff_nuc_ranges[, 2]) + as.integer(diff_nuc_ranges[, 4])
    )
)
diff_nuc_ranges <- resize(diff_nuc_ranges, 1000, fix = "center")

# Find regions contaning differential nucleosomes as foreground
filter <- (diff_nuc$mean_diff_score < -1) & (diff_nuc$score_fdr < 0.01)
nuc_filt <- diff_nuc_ranges[filter]

# Overlap with de novo moif sites
nuc_filt <- subsetByOverlaps(nuc_filt, TF_sites)
```

```r
# Keep motif sites around differential nucleosomes
TF_sites <- subsetByOverlaps(TF_sites, nuc_filt)

# For each age group, calculate the TF binding score for each motif site
TFBS_scores <- list()
for(age in c("Old", "Young")){
    if(age == "Old"){
        bigwig <- old_TFBS_bw
    }else{
        bigwig <- young_TFBS_bw
    }
    ov <- findOverlaps(TF_sites, bigwig)
    ovMat <- Matrix::sparseMatrix(
        i = ov@from,
        j = ov@to,
        x = 1, dims = c(length(TF_sites), length(bigwig))
    )
    scores <- (ovMat %*% bigwig$score) 
    TFBS_scores[[age]] <- scores[, 1] / width(TF_sites) # Convert a n-by-1 matrix to vector
}

# Get the difference across aging
diffScores <- TFBS_scores[["Old"]] - TFBS_scores[["Young"]]
```

```r
qplot(TFBS_scores[["Young"]], diffScores)
```
