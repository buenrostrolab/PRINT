# TF motif mutation experiments (Analysis)

```python
# First extract genomic DNA sequence from the target locus
chr = 'chr19'
left, right = 15924673, 15925472
center = (left + right) // 2
flank = 920
seq = scp.genome.mm10.fetch_seq(chr, 
                               center-flank, 
                               center+flank)
seq2 = list(seq)

# Next we find TF motif sites in the target locus
tp_record = TFBS[(TFBS[3] == 'Atf4') & (TFBS[0] == chr) & (TFBS[1] >= center-flank) & (TFBS[2] <= center+flank)]
ls, rs = np.array(tp_record[1]), np.array(tp_record[2])

# Replace motif sequence with random ACGT
for l,r in zip(ls, rs):
    l, r = int(l), int(r)
    size = r-l
    seq2[l-(center-flank): r-(center-flank)] = random.choices('ACGT', k=r-l)
    p = int(random.random() *  (1840-size))
    
# One-hote encode the mutated sequences
seq2 = ''.join(seq2)
seq = scp.utils.DNA_one_hot(seq)[None].float().cuda()
seq2 = scp.utils.DNA_one_hot(seq2)[None].float().cuda()

# Predicted footprints for mutated and wildtype sequences
pred = {}
pred_mut = {}
attrs = {}
for name in models:
    pred[name] = 0
    pred_mut[name] = 0
    attrs[name] = 0
    for m in models[name]:
        m = m.cuda()
        pred[name] += scp.utils.zscore2pval(m(seq)[0][0].detach().cpu().numpy())
        pred_mut[name] += scp.utils.zscore2pval(m(seq2)[0][0].detach().cpu().numpy())
        m = m.cpu()
        wrapper = JustSumWrapper(m, specific_pos=torch.arange(500, 600), 
                                 nth_output=torch.arange(45, 75), res=1, threshold=0.301).cuda()
        attributions = calculate_attributions(wrapper,
                                              X=seq.cpu(),
                                              n_shuffles=20,
                                              method='shap_hypo', verbose=False)
        attrs[name] += (attributions.cpu() * seq.cpu())[0].detach().cpu().numpy()[:, 520:-520]
        wrapper = wrapper.cpu()
    pred[name] /= 5
    pred_mut[name] /= 5
    attrs[name] /= 5
```

```python
pickle.dump([pred, pred_mut, attrs], open(os.path.join(model_dir, "Fig4_data1.pkl"), 'wb'))
```

## Identify age-associated TFs

```python
from adjustText import adjust_text

# Load TF binding scores
TFBS_path = "/home/yanhu/data/PRINT/multiScaleFootprinting/data/mHSCAging10xV3/TFBS_pred.tsv"
TFBS_scores = pd.read_csv(TFBS_path, sep = "\t")
TFs = np.unique(TFBS_scores["TF"].values)
```

```python
TFBS_scores.columns
```

```python
plot_dir = '/data/yanhu/PRINT/multiScaleFootprinting/data/mHSCAging10xV3/plots/sequence_model/'
```

```python
obs_diff = {"count":{}, "foot":{}}
for TF in tqdm(TFs):

    # Use fine-tuned TF binding scores either derived from sequence-to-footprint attribution scores
    # or from sequence-to-accessibility (count) attribution scores
    for mode in ["count", "foot"]:

        # Extract TF binding scores for each age group
        TF_scores = TFBS_scores[TFBS_scores["TF"].values == TF]
        old_TF_scores = TF_scores["seqTF_" + mode + "_old"].values
        young_TF_scores = TF_scores["seqTF_" + mode + "_young"].values

        # Calculate difference in mean TF binding scores and pvalue
        diff_scores = old_TF_scores - young_TF_scores
        stat = ss.ttest_1samp(diff_scores, popmean=0).statistic
        obs_diff[mode][TF] = {"mean_diff":np.mean(diff_scores), "stat":stat}

count_diff = np.array([obs_diff["count"][TF]["mean_diff"] for TF in TFs])
fp_diff = np.array([obs_diff["foot"][TF]["mean_diff"] for TF in TFs])
count_stat = np.array([obs_diff["count"][TF]["stat"] for TF in TFs])
fp_stat = np.array([obs_diff["foot"][TF]["stat"] for TF in TFs])
```

```python
motif_score_path = "../../data/mHSCAging10xV3/motifScores.rds"
motif_scores = pyreadr.read_r(motif_score_path)[None]
pbulk_age = np.array([re.split("_", i)[0] for i in motif_scores.index])
```

```python
chromVAR_diff = []
chromVAR_stat = []
for TF in tqdm(TFs):
    young_scores = motif_scores.loc[pbulk_age == "Young", TF]
    old_scores = motif_scores.loc[pbulk_age == "Old", TF]
    chromVAR_diff.append(np.mean(old_scores) - np.mean(young_scores))
    t_test = ss.ttest_ind(old_scores, young_scores)
    chromVAR_stat.append(t_test.statistic)
```

```python
scores = np.array(fp_stat)
plot_order = np.argsort(-scores)
plt.figure(dpi = 200, figsize=(8, 4))
plt.scatter(np.arange(len(scores)), scores[plot_order], s=2)
label_TFs = ["Atf4", "Atf6b", "Bcl11a", "Clock", "E2f1", "Egr1",
              "Ets1", "Fli1", "Jun", "Klf6", "Mycn", "Nfe2", "Nfe2l2",
              "Pbx3", "Smad1", "Spi1", "Tfec", "Xbp1"]
label_TFs = label_TFs + [TFs.tolist()[i] for i in plot_order[:10]]
for TF in label_TFs:
    i = list(TFs[plot_order]).index(TF)
    labels.append(plt.text(i, scores[plot_order][i], TF))
adjust_text(labels, arrowprops=dict(arrowstyle="-", color='black'), avoid_self=False)
plt.xlabel("Rank")
plt.ylabel("T statistic")
plt.show()
```
