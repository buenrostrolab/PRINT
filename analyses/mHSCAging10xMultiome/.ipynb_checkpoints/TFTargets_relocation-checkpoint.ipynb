```r
# If running in Rstudio, set the working directory to current path
if (Sys.getenv("RSTUDIO") == "1"){
  setwd(dirname(rstudioapi::getActiveDocumentContext()$path))
}

myPath <- .libPaths()
myPath <- c(myPath,'/packages')
.libPaths(myPath)

source("../../code/utils.R")
source("../../code/getCounts.R")
source("../../code/getBias.R")
source("../../code/getFootprints.R")
source("../../code/visualization.R")
source("../../code/getTFBS.R")
source("../../code/getTargetPathway.R")
library("ComplexHeatmap")

###################
# Load input data #
###################

# Load region ranges
regions <- readRDS("../../data/mHSCAging10xV3/regionRanges.rds")

# Load pathway gene sets from hypeR
library(hypeR)
c5GO <- msigdb_gsets(species = "Mus musculus","C5","BP",clean = TRUE)$genesets
names(c5GO) <- stringr::str_replace_all(names(c5GO), " ",  "_")

# Get CRE to gene mapping
TSS <- FigR::mm10TSSRanges
TSSCREOv <- findOverlaps(resize(TSS, 1000, fix = "center"), regions)
CREGeneMapping <- data.frame(
  Gene = as.character(TSS$gene_name)[TSSCREOv@from],
  CREInd = TSSCREOv@to,
  CRE = as.character(regions)[TSSCREOv@to],
  Region = as.character(regions[TSSCREOv@to])
)
```

```r
##########################
# Get TF motif positions #
##########################

# Load PWM data
cisBPMotifs <- readRDS("../../data/shared/cisBP_mouse_pwms_2021.rds")

# Find motif matches for all TFs
motifPath <- "../../data/mHSCAging10xV3/TFMotifRanges.rds"
if(!file.exists(motifPath)){
  TFMotifRanges <- pbmcapply::pbmclapply(
    names(cisBPMotifs),
    function(TF){
      motifmatchr::matchMotifs(cisBPMotifs[TF], 
                               regions, 
                               genome = "mm10",
                               out = "positions",
                               p.cutoff = 1e-4)[[1]]
    },
    mc.cores = 16
  )
  names(TFMotifRanges) <- names(cisBPMotifs)
  saveRDS(TFMotifRanges, motifPath)
}else{
  TFMotifRanges <- readRDS(motifPath)
}
```

```r
TFBS_path <- "/home/yanhu/data/PRINT/multiScaleFootprinting/data/mHSCAging10xV3/TFBS_pred.tsv"
TFBS <- data.table::fread(TFBS_path, sep = "\t", nThread = 8, showProgress = T)
```

```r
head(TFBS)
```

```r
# Retrieve positions of candidate TF binding sites (resize to 1 bp)
TFBSRanges <- GRanges(
    seqnames = TFBS$chrom, 
    ranges = IRanges(start = TFBS$start, end = TFBS$end))
TFBSRanges <- resize(TFBSRanges, 1, fix = "center")

# Generate a position-by-sample matrix of TF bindings scores
TFBindingSE <- SummarizedExperiment(
  assays = list(TFBS = as.matrix(TFBS$seqTF_foot_young)),
  rowRanges = TFBSRanges
)
```

```r
# Select top variant TFs
aging_diff = TFBS$seqTF_foot_old - TFBS$seqTF_foot_young
relativeVariance <- pbmcapply::pbmcmapply(
    function(TF){
        TF_site_inds <- TFBS$TF == TF
        sd(aging_diff[TF_site_inds]) / mean(TFBS$seqTF_foot_young)
    },
    unique(TFBS$TF),
    mc.cores = 16
)

topVariantTFs <- sort(relativeVariance, decreasing=T)[1:30]
topVariantTFs
```

```r
# Generate a position-by-sample matrix of differential TF bindings scores
TFBindingSE <- SummarizedExperiment(
  assays = list(TFBS = -as.matrix(TFBS$seqTF_foot_old - TFBS$seqTF_foot_young)),
  rowRanges = TFBSRanges
)
```

```r
enrichment <- mapTargets(
  TFMotifSites = TFMotifRanges[names(topVariantTFs)], 
  siteSE = TFBindingSE,
  regionGeneCorr = CREGeneMapping,
  geneSets = c5GO,
  threshold = 0.9,
  genome = "mm10",
  returnTargets = T
)
```

```r
saveRDS(enrichment, "../../data/mHSCAging10xV3/regulatorEnrichment_variant_TFs_lost.rds")
```

```r
enrichment_gained <- readRDS("../../data/mHSCAging10xV3/regulatorEnrichment_variant_TFs_gained.rds")
enrichment_lost <- readRDS("../../data/mHSCAging10xV3/regulatorEnrichment_variant_TFs_lost.rds")
```

```r
names(enrichment_lost)
```

```r
TF <- "Clock"
upEnrich <- enrichment_gained[[TF]]$enrichment
upEnrich <- upEnrich[upEnrich$fdrs < 0.25, ]
downEnrich <- enrichment_lost[[TF]]$enrichment
downEnrich <- downEnrich[downEnrich$fdrs < 0.25, ]
```

```r
options(repr.plot.width=10, repr.plot.height=8)
plotData <- rbind(upEnrich, downEnrich[dim(downEnrich)[1]:1, ])
plotData$logFDR <- -log10(plotData$fdr)
plotData$group <- c(rep("Gained", dim(upEnrich)[1]), rep("Lost", dim(downEnrich)[1]))
plotData$pathway <- factor(plotData$pathway, levels = plotData$pathway)
ggplot(plotData) +
    geom_bar(aes(x=logFDR, y=pathway, fill=group), stat = "identity", width = 0.5)
```
