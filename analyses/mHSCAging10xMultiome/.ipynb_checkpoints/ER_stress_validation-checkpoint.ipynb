```python
import numpy as np
import pandas as pd
import os
import re
import matplotlib.pyplot as plt
import pyBigWig as pw
import tqdm
import multiprocessing as mp
import sys
sys.path.append("/home/yanhu/data/PRINT/multiScaleFootprinting/code")
import scprinter as scp
```

```python
def fetch_bw(args):
    TFBS, bw = args
    chroms, starts, ends = np.array(TFBS['chrom']),np.array(TFBS['start']),np.array(TFBS['end'])
    res_all = {}
    with pw.open(bw, 'r') as f:
        for chrom in tqdm.tqdm(scp.genome.mm10.chrom_sizes):
            if chrom == 'chrY':
                continue
            res_all[chrom] = f.values(chrom, 0, scp.genome.mm10.chrom_sizes[chrom], numpy=True)
        
    vs = []
    for chr, left, right in zip(tqdm.tqdm(chroms, mininterval=1), starts, ends):
        vs.append(np.nanmean(res_all[chr][left:right]))
    return vs
```

```python
# Load aging TF binding scores 
print("Loading aging TFBS scores")
TFBS_path = "/home/yanhu/data/PRINT/multiScaleFootprinting/data/mHSCAging10xV3/TFBS_pred.tsv"
TFBS = pd.read_csv(TFBS_path, sep = "\t")
TFs = np.unique(TFBS["TF"].values)
TFBS = TFBS[["chrom", "start", "end", "TF", "seqTF_count_old",
             "seqTF_count_young", "seqTF_foot_old", "seqTF_foot_young"]]
```

```python
# Multi-process loading of TF binding scores from the LSK tunicamycin experiment
Tm_TFBS_dir = "/home/yanhu/data/PRINT/multiScaleFootprinting/data/LSKTunicamycin/seqPrint_TFBS_scores"
args = [[TFBS, os.path.join(Tm_TFBS_dir, "LSK_" + condition + "_TFBS.bigwig")]
       for condition in ["control", "treated"]]
n_jobs = 2
with mp.Pool(n_jobs) as pool:
    Tm_TFBS_scores = list(pool.imap(fetch_bw, args))
```

```python
TFBS["seqTF_LSK_control"] = Tm_TFBS_scores[0]
TFBS["seqTF_LSK_treated"] = Tm_TFBS_scores[1]
```

```python
aging_diff = TFBS["seqTF_foot_old"] - TFBS["seqTF_foot_young"]
Tm_diff = TFBS["seqTF_LSK_treated"] - TFBS["seqTF_LSK_control"]

z_scores = []
for TF in tqdm.tqdm(TFs):
    TF_site_inds = TFBS["TF"] == TF
    aging_z = np.mean(aging_diff[TF_site_inds]) / np.std(aging_diff[TF_site_inds])
    Tm_z = np.mean(Tm_diff[TF_site_inds]) / np.std(Tm_diff[TF_site_inds])
    z_scores.append([aging_z, Tm_z])
```

```python
z_scores = pd.DataFrame(z_scores, index=TFs, columns=["aging", "Tm"])
```
