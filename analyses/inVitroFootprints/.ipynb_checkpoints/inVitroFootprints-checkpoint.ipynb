```r
library(rtracklayer)
library(ggplot2)
```

```r
# Load TOBIAS footprint score bigwigs
resultDir <- "/home/yanhu/data/PRINT/multiScaleFootprinting/data/inVitroFootprints/240607_TOBIAS/"
sampleTable <- list(
    "MYC/MAX" = "FT5_5_S185",
    "Control" = "FT6_11_S190",
    "CEBPA" = "FT7_11_S200"
)
bigwigs <- lapply(
    names(sampleTable),
    function(TF){
        rtracklayer::import.bw(
            paste0(resultDir, sampleTable[[TF]], 
                   "_001.st.rmdup.flt_corrected.footprints.bw"))
    }
)
names(bigwigs) <- names(sampleTable)
```

```r
# Load TF motif sites
motifBed <- list(
    "MYC/MAX" = "/home/yanhu/data/PRINT/multiScaleFootprinting/data/inVitroFootprints/MycMaxMotifSites.bed",
    "CEBPA" = "/home/yanhu/data/PRINT/multiScaleFootprinting/data/inVitroFootprints/CebpaMotifSites.bed"
)
motifSites <- lapply(
    names(motifBed),
    function(TF){
        bed <- read.table(motifBed[[TF]], sep = ",", header = T)
        GRanges(seqnames = bed[, 1], ranges = IRanges(start = bed[, 2], end = bed[, 3]))
    }
)
names(motifSites) <- names(motifBed)
```

```r
TF <- "CEBPA"

# Find positions overlapping with motifs
ov <- findOverlaps(bigwigs[[TF]], motifSites[[TF]])

# Get footprint scores at motif sites in the presence of TFs
motifTFScores <- data.frame(
    condition = paste0("Motif sites\n", paste(TF, "(+)")),
    score = bigwigs[[TF]][ov@from]$score
)

# Find background positions not overlapping with motifs
bgTFScores <- data.frame(
    condition = paste0("Background sites\n", paste(TF, "(+)")),
    score = bigwigs[[TF]][-ov@from]$score
)

# Get footprint scores at motif sites in the absense of TFs
ov <- findOverlaps(bigwigs[["Control"]], motifSites[[TF]])
motifControlScores <- data.frame(
    condition = paste0("Motif sites\n", paste(TF, "(-)")),
    score = bigwigs[["Control"]][ov@from]$score
)

# Find background positions not overlapping with motifs
bgControlScores <- data.frame(
    condition = paste0("Background sites\n", paste(TF, "(-)")),
    score = bigwigs[["Control"]][-ov@from]$score
)

scores <- Reduce(rbind, list(motifTFScores, bgTFScores, 
                             motifControlScores, bgControlScores))

# Fix condition order
scores$condition = factor(
    scores$condition, 
    levels = c(
        paste0("Motif sites\n", paste(TF, "(-)")),
        paste0("Motif sites\n", paste(TF, "(+)")),
        paste0("Background sites\n", paste(TF, "(-)")),
        paste0("Background sites\n", paste(TF, "(+)"))
    )
)
```

```r
ggplot(scores, aes(x=condition, y = score)) +
    stat_boxplot(geom ='errorbar') + 
    geom_boxplot() + theme_classic() 
```
