### Continuing TFBS deletion analysis with updated list and idea to do correlation matrix

```python
import pickle
import os
import warnings
import seaborn as sns
warnings.simplefilter(action='ignore', category=FutureWarning)
import matplotlib as mpl
mpl.rcParams['pdf.fonttype'] = 42
import numpy as np
from matplotlib import pyplot as plt
from scipy import cluster
import pandas as pd
from tqdm import tqdm
from matplotlib.colors import LinearSegmentedColormap
import pyranges as pr
wd = '/home/yanhu/data/PRINT/multiScaleFootprinting/data/sequenceModels/TF_clustering'
```

```python
# Load dictionary storing results for each TF
tf_dict = pickle.load(open(os.path.join(wd, "tf_delta_v2.pkl"), 'rb'))
tf_list = list(tf_dict.keys())

# Select delta footprint scores from representative scales
scales = [10, 20, 30, 50]
delta_matrix = [[tf_delta[scale - 2, :] for tf, tf_delta in tf_dict.items()] for scale in scales]
delta_matrix = [(x - np.mean(x)) / np.std(x) for x in delta_matrix]
delta_matrix = np.hstack(delta_matrix)

# Calculate correlation matrix
corr_matrix = np.corrcoef(delta_matrix)
```

```python
# Cluster and visualize the correlation matrix
corr_df = pd.DataFrame(corr_matrix, index=tf_list, columns=tf_list)
colors = ['#3361A5', '#248AF3', '#14B3FF', '#88CEEF', '#C1D5DC', '#EAD397', '#FDB31A', '#E42A2A', '#A31D1D']
colors = [[i/(len(colors)-1), colors[i]] for i in range(len(colors))]
cmap = LinearSegmentedColormap.from_list("", colors)
g = sns.clustermap(corr_df, cmap=cmap, figsize=(100, 100), yticklabels=True, vmax=1, vmin=-1)

# Mask lower half triangle
mask = np.tril(np.ones_like(corr_matrix))
values = g.ax_heatmap.collections[0].get_array().reshape(corr_matrix.shape)
masked_values = np.ma.array(values, mask=mask)
g.ax_heatmap.collections[0].set_array(masked_values)

# Save plot
plt.savefig(os.path.join(wd, "plots", "TF_clustermap.pdf"))
plt.show()
```

```python
selected_TFs = ["ZKSCAN1", "JUN", "CEBPD", "YY1", "NFIB"]
for TF in selected_TFs:
    sns.heatmap(np.flip(tf_dict[TF], axis=0), cmap="RdBu", vmin=-0.5, vmax=0.5)
    
    # Save plot
    plt.savefig(os.path.join(wd, "plots", TF + "_delta.pdf"))
    plt.show()
```

### Compare to TF clustering using binding site similarity

```python
# Retrieve TF binding sites of each TF
tfbs = pd.read_csv(os.path.join(wd, "filteredTFBS_cat.tsv"), sep="\t")
tfbs_dict = {tf:tfbs[tfbs["TF"] == tf] for tf in tf_list}
tfbs_dict = {}
for tf in tf_list:
    tfbs_ = tfbs[tfbs["TF"] == tf]
    tfbs_.columns = ["Chromosome", "Start", "End", "Strand", "TF"]
    tfbs_ = pr.PyRanges(tfbs_)
    tfbs_dict[tf] = tfbs_
```

```python
# Calculate pairwise jaccard similarity of TF binding sites among TFs
n_tfs = len(tf_list)
jaccard_matrix = -np.ones((n_tfs, n_tfs)) # Use -1 as "not yet computed flag"
for tf_ind_1 in tqdm(range(n_tfs)):
    for tf_ind_2 in range(n_tfs):
        if jaccard_matrix[tf_ind_2, tf_ind_1] >= 0:
            # The matrix should be symmetric
            jaccard_matrix[tf_ind_1, tf_ind_2] = jaccard_matrix[tf_ind_2, tf_ind_1]
        else:
            tf_1, tf_2 = tf_list[tf_ind_1], tf_list[tf_ind_2]
            pair_ranges = [tfbs_dict[tf_1], tfbs_dict[tf_2]]
            intersect = len(pair_ranges[0].intersect(pair_ranges[1], how="first"))
            union = len(pair_ranges[0]) +  len(pair_ranges[1]) - intersect
            jaccard = intersect / union
            jaccard_matrix[tf_ind_1, tf_ind_2] = jaccard
```

```python
# Cluster and visualize the correlation matrix
jaccard_df = pd.DataFrame(jaccard_matrix, index=tf_list, columns=tf_list)
colors = ['#3361A5', '#248AF3', '#14B3FF', '#88CEEF', '#C1D5DC', '#EAD397', '#FDB31A', '#E42A2A', '#A31D1D']
colors = [[i/(len(colors)-1), colors[i]] for i in range(len(colors))]
cmap = LinearSegmentedColormap.from_list("", colors)
g = sns.clustermap(jaccard_df, cmap=cmap, figsize=(100, 100), yticklabels=True, vmax=0.05)

# Mask lower half triangle
mask = np.tril(np.ones_like(corr_matrix))
values = g.ax_heatmap.collections[0].get_array().reshape(corr_matrix.shape)
masked_values = np.ma.array(values, mask=mask)
g.ax_heatmap.collections[0].set_array(masked_values)

# Save plot
plt.savefig(os.path.join(wd, "plots", "TF_clustermap_TFBS_jaccard.pdf"))
plt.show()
```
