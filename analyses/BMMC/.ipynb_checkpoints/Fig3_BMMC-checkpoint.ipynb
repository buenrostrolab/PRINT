```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pickle
import sys
import os
import re
import matplotlib
import pyBigWig
sys.path.append("/home/yanhu/data/PRINT/multiScaleFootprinting/code")
import scprinter as scp
matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42

data_dir = "/home/yanhu/data/PRINT/multiScaleFootprinting/data/BMMC/"
BMMC_all_result_path_1 = os.path.join(data_dir, "Bulk_LoRA_example_results.pkl")
BMMC_all_result_path_2 = os.path.join(data_dir, "Bulk_LoRA_example_results_new.pkl")
BMMC_Ery_result_path = os.path.join(data_dir, "Ery_LoRA_example.pkl")
```

```python
# Define functions for smoothing

def strided_axis0(a, L):
    # Store the shape and strides info
    shp = a.shape
    s  = a.strides

    # Compute length of output array along the first axis
    nd0 = shp[0]-L+1

    # Setup shape and strides for use with np.lib.stride_tricks.as_strided
    # and get (n+1) dim output array
    shp_in = (nd0,L)+shp[1:]
    strd_in = (s[0],) + s
    return np.lib.stride_tricks.as_strided(a, shape=shp_in, strides=strd_in)

def smooth(proj_attrs):
    proj_attrs = strided_axis0(proj_attrs, 3).mean(axis=1)
    proj_attrs = np.transpose(proj_attrs, (1, 0))
    proj_attrs = strided_axis0(proj_attrs, 5).mean(axis=1)
    proj_attrs = np.transpose(proj_attrs, (1, 0))
    return proj_attrs
```

```python
# Load pre-computed results for visualization
with open(BMMC_all_result_path_1, "rb") as f:
    results = pickle.load(f)
```

```python
results.keys()
```

```python
for region in results.keys():

    radius = 400
    chr, start, end = re.split("[:-]", region)
    center = (int(start) + int(end)) // 2
    start, end = center - radius, center + radius

    # Extract predicted TF binding score and nucleosome score matrix
    pred_TF_matrix, pred_nuc_matrix = results[region]

    pred_TF_matrix = smooth(pred_TF_matrix)
    sns.heatmap(pred_TF_matrix, cmap="RdBu_r", 
                vmin=0, vmax=0.75)
    region = chr + ":" + str(start) + "-" + str(end)
    plt.savefig(os.path.join(data_dir, "plots", "seq2PRINT_TFBS", region + "_seq2PRINT_TFBS.png"))
    plt.show()

    sns.heatmap(pred_nuc_matrix, cmap="Blues")
    plt.show()
```

### Visaulize accessibility tracks for specific cell types

```python
plot_dir = "/home/yanhu/data/PRINT/multiScaleFootprinting/data/BMMC/plots/ATAC_tracks"
if not os.path.exists(plot_dir):
    os.mkdir(plot_dir)
```

```python
chr = "chr11"
region_start = 47364152
region_end = 47392951
CRE_start = 47378111
CRE_end = 47378911

# Load bigwig file of ATAC coverage
plt.figure(figsize=(18, 3), dpi=300)
cell_types = ["early-Ery", "CD14mono", "CLP", "MemoryB"]
coverage_list = []
for cell_type in cell_types:
    bigwig_path = "/home/yanhu/data/PRINT_rev/BMMC/peak_call/" + cell_type + "_treat_pileup.bw"
    bw = pyBigWig.open(bigwig_path, "r")
    coverage_list.append(bw.values(chr, region_start, region_end))

colors = ["#c0b0c0", "#c09090", "#90c090", "#90c0c0"]
max_coverage = [np.max(i[int(len(i)/2)]) for i in coverage_list]
for i in np.argsort(-np.array(max_coverage)):
    plt.fill_between(np.arange(region_start, region_end), coverage_list[i], 
                     label=cell_types[i], color=colors[i])
plt.axvline(x = CRE_start, color = 'r', linestyle = 'dashed') 
plt.axvline(x = CRE_end, color = 'r', linestyle = 'dashed') 
plt.legend()
plt.savefig(plot_dir + "/" + "_".join([chr, str(region_start), str(region_end), 
                                 "_".join(cell_types)]) + ".pdf")
```
